```python
import requests
import pandas as pd

# Empty list to store earthquake records
all_records = []

# API URL
url = "https://earthquake.usgs.gov/fdsnws/event/1/query"

# Year range
start_year = 2020
end_year = 2025

# Loop through years
for year in range(start_year, end_year + 1):

    # Loop through months
    for month in range(1, 13):

        # Create start date
        start_date = f"{year}-{month:02d}-01"

        # Create end date
        if month == 12:
            end_date = f"{year + 1}-01-01"
        else:
            end_date = f"{year}-{month + 1:02d}-01"

        # API parameters
        params = {
            "format": "geojson",
            "starttime": start_date,
            "endtime": end_date,
            "minmagnitude": 3
        }

        # Send API request
        response = requests.get(url, params=params)

        # Check request status
        if response.status_code != 200:
            print(f"Failed for {start_date}: {response.text[:200]}")
            continue

        # Convert response into JSON
        try:
            data = response.json()

        except Exception as e:
            print(f"JSON error for {start_date}: {e}")
            continue

        # Extract earthquake records
        for f in data["features"]:

            # Earthquake properties
            p = f["properties"]

            # Coordinates [longitude, latitude, depth]
            g = f["geometry"]["coordinates"]

            # Store earthquake data
            all_records.append({

                "id": f.get("id"),

                "time": pd.to_datetime(
                    p.get("time"), unit="ms"
                ),

                "updated": pd.to_datetime(
                    p.get("updated"), unit="ms"
                ),

                "latitude": g[1] if g else None,
                "longitude": g[0] if g else None,
                "depth_km": g[2] if g else None,

                "mag": p.get("mag"),
                "magType": p.get("magType"),
                "place": p.get("place"),
                "status": p.get("status"),
                "tsunami": p.get("tsunami"),
                "sig": p.get("sig"),
                "net": p.get("net"),
                "nst": p.get("nst"),
                "dmin": p.get("dmin"),
                "rms": p.get("rms"),
                "gap": p.get("gap"),
                "magError": p.get("magError"),
                "depthError": p.get("depthError"),
                "magNst": p.get("magNst"),
                "locationSource": p.get("locationSource"),
                "magSource": p.get("magSource"),
                "types": p.get("types"),
                "ids": p.get("ids"),
                "sources": p.get("sources"),
                "type": p.get("type")

            })

# Convert list into DataFrame
df = pd.DataFrame(all_records)

```

```python
import pandas as pd

# Extract country/region from place column using Regex
df["country"] = df["place"].str.extract(r",\s*([^,]+)$")

# Convert alert column to lowercase (if column exists)
if "alert" in df.columns:
    df["alert"] = df["alert"].str.lower()

# List of text columns to clean
text_columns = [
    "magType",
    "status",
    "type",
    "net",
    "sources",
    "types"
]

# Clean string columns
for col in text_columns:

    # Check whether column exists
    if col in df.columns:

        # Convert values to string
        df[col] = df[col].astype(str)

        # Remove extra spaces
        df[col] = df[col].str.strip()

        # Convert text to lowercase
        df[col] = df[col].str.lower()

```

```python
# Display cleaned dataset
print(df.head())
```

```python
import pandas as pd

# List of numeric columns
numeric_columns = [
    "mag",
    "depth_km",
    "nst",
    "dmin",
    "rms",
    "gap",
    "magError",
    "depthError",
    "magNst",
    "sig"
]

# Convert columns to numeric datatype
for col in numeric_columns:

    # Check whether column exists
    if col in df.columns:

        # Convert values into numeric format
        df[col] = pd.to_numeric(
            df[col],
            errors="coerce"
        )

        # Fill missing values with median
        df[col] = df[col].fillna(
            df[col].median()
        )

# Display cleaned numeric data
print(df[numeric_columns].head())
```

```python
import pandas as pd

# Extract date-related columns from time
df["year"] = df["time"].dt.year

df["month"] = df["time"].dt.month

df["day"] = df["time"].dt.day

df["day_of_week"] = df["time"].dt.day_name()

# Create shallow/deep earthquake flag
df["depth_category"] = df["depth_km"].apply(
    lambda x: "shallow" if x < 70 else "deep"
)

# Create earthquake strength category
def classify_magnitude(mag):

    if mag >= 7:
        return "destructive"

    elif mag >= 5:
        return "strong"

    else:
        return "moderate"

# Apply magnitude classification
df["magnitude_category"] = df["mag"].apply(
    classify_magnitude
)

# Display updated dataset
print(df.head())
```

```python
%%writefile app.py

import streamlit as st

# ---------------------------------------------------
# PAGE CONFIG (MUST BE FIRST STREAMLIT COMMAND)
# ---------------------------------------------------
st.set_page_config(
    page_title="Earthquake Data Analysis",
    layout="wide"
)

import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt

# ---------------------------------------------------
# DATABASE CONNECTION
# ---------------------------------------------------
@st.cache_resource
def get_engine():
    return create_engine(
        "mysql+pymysql://root:Anitha99@localhost/Earthquake_Project"
    )

engine = get_engine()


def get_data(query, params=None):
    return pd.read_sql(query, engine, params=params)


# ---------------------------------------------------
# SIDEBAR
# ---------------------------------------------------
st.sidebar.title("🌍 Navigation")

page = st.sidebar.radio(
    "Go To",
    (
        "Project Introduction",
        "Earthquake Visualization",
        "SQL Analysis",
        "Creator Info"
    )
)

# ===================================================
# PAGE 1
# ===================================================
if page == "Project Introduction":

    st.title("🌍 Earthquake Data Analysis Dashboard")

    st.markdown("""
This dashboard analyzes earthquake data stored in MySQL.

### Features

- Earthquake filtering
- Magnitude analysis
- Depth analysis
- Location Map
- SQL Insights
- Download filtered dataset
""")

# ===================================================
# PAGE 2
# ===================================================
elif page == "Earthquake Visualization":

    st.title("🌍 Earthquake Visualization")

    min_mag = st.slider(
        "Minimum Magnitude",
        0.0,
        10.0,
        2.5
    )

    max_mag = st.slider(
        "Maximum Magnitude",
        0.0,
        10.0,
        7.0
    )

    start_date = st.date_input("Start Date")
    end_date = st.date_input("End Date")

    place_search = st.text_input("Search Place")

    query = """
    SELECT *
    FROM earthquakes
    WHERE mag BETWEEN %s AND %s
      AND DATE(time) BETWEEN %s AND %s
      AND place LIKE %s
    ORDER BY time DESC
    """
    df = get_data(
        query,
        (
            min_mag,
            max_mag,
            start_date,
            end_date,
            f"%{place_search}%"
        )
    )

    if not df.empty:

        df["time"] = pd.to_datetime(df["time"])

        df = df.dropna(
            subset=[
                "latitude",
                "longitude",
                "mag",
                "depth"
            ]
        )

        st.subheader("Filtered Data")

        st.dataframe(
            df,
            use_container_width=True
        )

        csv = df.to_csv(index=False).encode("utf-8")

        st.download_button(
            "⬇ Download CSV",
            csv,
            "earthquakes.csv",
            "text/csv"
        )

        st.subheader("Earthquake Locations")

        st.map(df[["latitude", "longitude"]])

        # Magnitude Trend
        st.subheader("Magnitude Trend")

        fig, ax = plt.subplots()

        ax.plot(
            df["time"],
            df["mag"]
        )

        ax.set_xlabel("Time")
        ax.set_ylabel("Magnitude")

        plt.xticks(rotation=45)

        st.pyplot(fig)

        # Scatter Plot
        st.subheader("Depth vs Magnitude")

        fig2, ax2 = plt.subplots()

        ax2.scatter(
            df["depth"],
            df["mag"]
        )

        ax2.set_xlabel("Depth")
        ax2.set_ylabel("Magnitude")

        st.pyplot(fig2)

        # Metrics
        col1, col2, col3 = st.columns(3)

        col1.metric(
            "Total Records",
            len(df)
        )

        col2.metric(
            "Maximum Magnitude",
            round(df["mag"].max(), 2)
        )

        col3.metric(
            "Average Depth",
            round(df["depth"].mean(), 2)
        )

        # Top 10
        st.subheader("Top 10 Strongest Earthquakes")

        top10 = get_data("""
            SELECT place, mag
            FROM earthquakes
            ORDER BY mag DESC
            LIMIT 10
        """)

        st.dataframe(top10)

        st.bar_chart(
            top10.set_index("place")
        )

    else:
        st.warning("No earthquake data found.")

# ===================================================
# PAGE 3
# ===================================================
elif page == "SQL Analysis":

    st.title("📊 SQL Analysis")

    queries = {

        "Top 10 Strongest Earthquakes":
        """
        SELECT place, mag
        FROM earthquakes
        ORDER BY mag DESC
        LIMIT 10
        """,

        "Top 10 Deepest Earthquakes":
        """
        SELECT place, depth
        FROM earthquakes
        ORDER BY depth DESC
        LIMIT 10
        """,

        "Average Magnitude by Magnitude Type":
        """
        SELECT magType,
               AVG(mag) AS Average_Magnitude
        FROM earthquakes
        GROUP BY magType
        """,

        "Earthquakes Per Hour":
        """
        SELECT HOUR(time) AS Hour,
               COUNT(*) AS Total
        FROM earthquakes
        GROUP BY HOUR(time)
        ORDER BY Hour
        """,

        "Status Count":
        """
        SELECT status,
               COUNT(*) AS Total
        FROM earthquakes
        GROUP BY status
        """
    }

    selected_query = st.selectbox(
        "Choose SQL Query",
        list(queries.keys())
    )

    result = get_data(
        queries[selected_query]
    )

    st.dataframe(
        result,
        use_container_width=True
    )

# ===================================================
# PAGE 4
# ===================================================
elif page == "Creator Info":

    st.title("👩‍💻 Creator")

    st.write("""
### Developed By

**Name:** Vinodhini

### Skills

- Python
- SQL
- Streamlit
- Data Analysis
""")

    st.image(
        "https://via.placeholder.com/150",
        width=150
    )

# ---------------------------------------------------
# FOOTER
# ---------------------------------------------------
st.markdown("---")
st.caption("Earthquake Data Analysis Dashboard")
```
